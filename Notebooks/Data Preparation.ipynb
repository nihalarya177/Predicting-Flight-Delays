## More data conversions

### Function to convert the String time to datetime.time

```python
def format_hour(key):
    if pd.isnull(key):
        return np.nan
    else:
        if key == 2400: 
            key = 0
        key = "{0:04d}".format(int(key))
        hour = datetime.time(int(key[0:2]), int(key[2:4]))
        return hour
```

### Function that combines a date and time to produce a datetime.datetime

```python
def combine_date_hour(x):
    if pd.isnull(x[0]) or pd.isnull(x[1]):
        return np.nan
    else:
        return datetime.datetime.combine(x[0],x[1])
```

### Function that combines two columns of the DataFrame to create a datetime format

```python
def create_flight_time(df, col):
    list1 = []
    for index, cols in df[['Date',col]].iterrows():
        if pd.isnull(cols[1]):
            list1.append(np.nan)
        elif float(cols[1]) == 2400:
            cols[0] += datetime.timedelta(days=1)
            cols[1] = datetime.time(0,0)
            list1.append(combine_date_hour(cols))
        else : 
            cols[1] = format_hour(cols[1])
            list1.append(combine_date_hour(cols))
    return pd.Series(list1)
```

```python
DF['Arrival_Time'] = DF['Arrival_Time'].apply(format_hour)
DF['Departure_Time'] = DF['Departure_Time'].apply(format_hour)
```

## Creating the Delay Levels column based on our metric for delay levels

1. On Time or Small Delay (up to 15 minutes delay)
2. Medium Delay (15  45 minutes delay)
3. Large Delay (45 minutes delay)

```python
delay_type = lambda x:((0,1)[x > 15],2)[x > 45]
DF['Delay_Level'] = DF['Departure_Delay'].apply(delay_type)
```
